```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.linspace(0, 10, 11)
b = a ** 4
```

```python
x = np.arange(0,10)
y = 2 * x
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0][0].plot(x, y)
axes[1][1].plot(a,b)

fig.suptitle("Figure LEVEL", fontsize=16)
plt.tight_layout() # automatically prevent overlap
# fig.subplots_adjust() # customization

fig.savefig('subplots.png', bbox_inches='tight')
```
